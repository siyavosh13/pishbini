```python
import os
import pydicom
import numpy as np
from PIL import Image

input_dir = 'test_dataset'  
output_dir = 'processed_png'

os.makedirs(output_dir, exist_ok=True)

def convert_dicom_to_png(dicom_path, output_path):
    ds = pydicom.dcmread(dicom_path)
    image = ds.pixel_array.astype(float)

    
    scaled_image = (np.maximum(image, 0) / image.max()) * 255.0
    scaled_image = np.uint8(scaled_image)

    img = Image.fromarray(scaled_image)
    img.save(output_path)


for filename in os.listdir(input_dir):
    if filename.lower().endswith('.dcm'):
        dicom_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename.replace('.dcm', '.png'))
        convert_dicom_to_png(dicom_path, output_path)
```

```python
import cv2

def preprocess_image(image_path, output_path, size=(224, 224)):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    resized_image = cv2.resize(image, size)
    normalized_image = resized_image / 255.0  

    
    cv2.imwrite(output_path, (normalized_image * 255).astype(np.uint8))
```

```python
processed_folder = 'processed_png'
output_folder = 'processed_png'

os.makedirs(processed_folder, exist_ok=True)

for filename in os.listdir(output_folder):
    if filename.lower().endswith('.png'):
        image_path = os.path.join(output_folder, filename)
        output_path = os.path.join(processed_folder, filename)
        preprocess_image(image_path, output_path)
```

```python
import os
import cv2
import numpy as np


processed_folder = 'processed_png'


image_files = [f for f in os.listdir(processed_folder) if f.endswith('.png')]


X_train = []
for file in image_files:
    image_path = os.path.join(processed_folder, file)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (224, 224))
    image = image.astype('float32') / 255.0  # نرمال‌سازی به بازه 0 تا 1
    image = np.expand_dims(image, axis=-1)  # افزودن بعد کانال
    X_train.append(image)

X_train = np.array(X_train)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_autoencoder(input_shape=(224, 224, 1)):
    input_img = tf.keras.Input(shape=input_shape)

    # Encoder
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(encoded)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    decoded = layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = models.Model(input_img, decoded)
    return autoencoder

autoencoder = build_autoencoder()
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
autoencoder.fit(X_train, X_train,
                epochs=50,
                batch_size=32,
                shuffle=True,
                validation_split=0.1)
```

```python
import matplotlib.pyplot as plt


test_img = X_train[0]
test_img_input = np.expand_dims(test_img, axis=0)
reconstructed_img = autoencoder.predict(test_img_input)


plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title('تصویر اصلی')
plt.imshow(test_img.squeeze(), cmap='gray')

plt.subplot(1, 2, 2)
plt.title('تصویر بازسازی‌شده')
plt.imshow(reconstructed_img.squeeze(), cmap='gray')

plt.show()
```

```python
# محاسبه خطای بازسازی برای هر تصویر
reconstructions = autoencoder.predict(X_train)
reconstruction_errors = np.mean(np.abs(reconstructions - X_train), axis=(1, 2, 3))

# تعیین آستانه برای تشخیص ناهنجاری
threshold = np.mean(reconstruction_errors) + 2 * np.std(reconstruction_errors)

# تصاویر با خطای بازسازی بالاتر از آستانه به‌عنوان ناهنجار در نظر گرفته می‌شوند
anomalies = reconstruction_errors > threshold
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D

input_img = Input(shape=(224, 224, 1))

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

# Decoder
x = Conv2D(64, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
autoencoder.fit(X_train, X_train,
                epochs=50,
                batch_size=32,
                shuffle=True,
                validation_split=0.1)
```

```python
import numpy as np
import matplotlib.pyplot as plt


test_image = X_train[0]  


reconstructed = autoencoder.predict(np.expand_dims(test_image, axis=0))[0]


error = np.abs(test_image - reconstructed)


plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(test_image.squeeze(), cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Reconstructed')
plt.imshow(reconstructed.squeeze(), cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Reconstruction Error')
plt.imshow(error.squeeze(), cmap='hot')

plt.show()
```
